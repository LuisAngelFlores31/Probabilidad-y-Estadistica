```python

# Datos de entrada: Horas de estudio y calificaciones
horas_estudio = np.array([2, 3, 5, 4, 11, 5]).reshape(-1, 1)
calificaciones = np.array([20, 25, 34, 30, 40, 31])

# Crear y entrenar el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(horas_estudio, calificaciones)

# Predecir la calificación para 8 horas de estudio
horas_prediccion = np.array([[8]])
calificacion_predicha = modelo.predict(horas_prediccion)

print("La calificación esperada para 8 horas de estudio es:", calificacion_predicha[0])
print("Valor de la intercepción (b):", modelo.intercept_)
print("Valor de la pendiente (m):", modelo.coef_[0])

# Crear el diagrama de dispersión con la línea de regresión
plt.figure(figsize=(6,5))
plt.scatter(horas_estudio, calificaciones, color='blue', label='Datos')
plt.plot(horas_estudio, modelo.predict(horas_estudio), color='red', label='Línea de regresión')
plt.scatter(horas_prediccion, calificacion_predicha, color='green', marker='x', s=100, label='Predicción (8 horas, {:.2f})'.format(calificacion_predicha[0]))
plt.title('Pronóstico de calificaciones según horas de estudio')
plt.xlabel('Horas de estudio')
plt.ylabel('Calificación final')
plt.legend()
plt.grid(True)
plt.show()
```

##1.¿Explicar cuál es la interpretación del coeficiente de regresión en términos de la relación entre el tiempo de estudio y las calificaciones finales?

El coeficiente de la pendiente en la regresión lineal muestra cuánto puede incrementarse, en promedio, la calificación final de un alumno por cada hora adicional que dedique al estudio. Por ejemplo, si la pendiente tiene un valor de 3, esto indica que por cada hora extra de estudio, se espera que la calificación aumente aproximadamente 3 puntos

## 2. ¿Cómo se interpreta el coeficiente de correlación en términos de la fuerza y dirección de la relación entre las variables?

El coeficiente de correlación de Pearson, permite conocer qué tan fuerte y en qué dirección se relacionan las horas de estudio con las calificaciones. Si el valor está cerca de +1, significa que existe una relación positiva y fuerte: al aumentar las horas de estudio, también aumentan las calificaciones. Si el valor se aproxima a 0, la relación entre ambas variables es muy débil o inexistente. Por otro lado, si el valor se acerca a -1, indica una relación negativa fuerte, lo que significa que las calificaciones tienden a bajar.

## 3. ¿El modelo de regresión lineal es estadísticamente significativo?

Sí, el modelo es estadísticamente significativo si el coeficiente de correlación es alto, el R² es elevado y el valor p es menor a 0.05. Esto indica que la relación entre horas de estudio y calificación es real y no producto del azar

## 4. ¿Qué tan confiable es el modelo de regresión para predecir las calificaciones finales de los estudiantes basado en el tiempo de estudio?

La confiabilidad del modelo depende de qué tan alto sea el R² y el coeficiente de correlación. Si ambos son elevados, significa que el modelo puede predecir con bastante precisión las calificaciones a partir de las horas de estudio.
